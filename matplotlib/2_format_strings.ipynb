```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,49,47,46]
```

```python
plt.plot(x,y,'g+') #colour to green and point is + marker
```

```python
plt.plot(x,y,'--D') #Diamond
```

no format issue

```python
plt.plot(x,y,'--r*')
```

argument passing also possible

hexadecimal cominatio for color is also possible

```python
plt.plot(x,y, color='#008040', marker='D', linestyle='dashed',markersize=10)
```

alpha argument for transparency 0 to 1 scale

```python
plt.plot(x,y, color='#008040', alpha=0.5)
```
