```python
import matplotlib.pyplot as plt
```

#need only single dim array

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
```

```python
plt.hist(blood_sugar,bins=3,rwidth=0.8) # by default number of bins is set to 10, rwidth max value is 1
```

we can customise range using bins

```python
plt.hist(blood_sugar,bins=[80,100,125,150],rwidth=0.8,color='red',histtype='step',orientation='horizontal')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

pass list when multiple case

```python
plt.hist([blood_sugar_men,blood_sugar_women],
         bins=[80,100,125,150],
         rwidth=0.8,
         color=['r','g'],
         label=['men','women'])
plt.xlabel('blood sugar')
plt.ylabel('no of patients')
plt.title('blood sugar analysis')
plt.legend()
```
