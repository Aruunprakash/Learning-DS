```python
import matplotlib.pyplot as plt
```

```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

perfect circle

```python
plt.pie(exp_vals,labels=exp_labels)
plt.axis("equal")
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.1,autopct="%1.1f%%")
plt.show()
```

explode , startangle

```python
plt.axis('equal')
plt.pie(exp_vals,labels=exp_labels,radius=1.1,autopct="%1.1f%%",explode=[0,0.3,0,0,0],startangle=90)
plt.savefig("piechart.jpg", bbox_inches="tight", pad_inches=1, transparent=True)#save as png file
```
