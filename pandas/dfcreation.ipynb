adding value using converter

```python
def converting_cell(cell):
    if cell == 'n.a.':
        return "sam walton"
    return cell

def eps_cell(cell):
    if cell == 'not available':
        return None
    return cell
```

```python
df =pd.read_excel("stock_data.xlsx","Sheet1",converters={
    'people':converting_cell,
    'eps':eps_cell
})
df
```

write in excel

```python
df.to_excel('new_data.xlsx',sheet_name="stocks")
df
```

```python
df.to_excel('new_data.xlsx',sheet_name="stocks",startrow=1,startcol=2,index=False)
df
```

Write two dataframes to two separate sheets in excel

```python
df_stocks = pd.DataFrame({
    'tickers': ['GOOGL', 'WMT', 'MSFT'],
    'price': [845, 65, 64 ],
    'pe': [30.37, 14.26, 30.97],
    'eps': [27.82, 4.61, 2.12]
})

df_weather =  pd.DataFrame({
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'event': ['Rain', 'Sunny', 'Snow']
})
```

```python
with pd.ExcelWriter('stocks_weather.xlsx') as writer:
    df_stocks.to_excel(writer, sheet_name="stocks")
    df_weather.to_excel(writer, sheet_name="weather")
```
