```python
import pandas as pd
```

```python
df = pd.read_excel("survey.xls")
df
```

cross tab is under pandas not dataframe

```python
pd.crosstab(df.Nationality,df.Handedness)
```

```python
pd.crosstab(df.Sex,df.Handedness,margins=True) #3rd arg is margin
```

```python
pd.crosstab([df.Nationality,df.Sex],df.Handedness,margins=True)
```

```python
pd.crosstab(df.Nationality,df.Handedness,normalize='index') #gives pct distribution
```

```python
import numpy as np
pd.crosstab(df.Sex,df.Nationality,values=df.Age,aggfunc=np.average)
```
