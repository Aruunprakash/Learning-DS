PIVOT TABLE :- allows to summarize and aggregate

```python
df = pd.read_csv("weather2.csv")
df
```

```python
df.pivot_table(index="city",columns="date")
```

aggfunc

```python
df.pivot_table(index="city",columns="date",aggfunc="sum") #count,diff,mean,
```

margin gives avg also

```python
df.pivot_table(index="city",columns="date",margins=True)
```
