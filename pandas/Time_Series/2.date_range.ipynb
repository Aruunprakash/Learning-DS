```python
df = pd.read_csv("aapl_nodate.csv")
df.info
```

```python
rng = pd.date_range(start="6/1/2017",periods=len(df),freq="B") #B means business days
rng
```

```python
df.set_index(rng,inplace=True)
```

```python
%matplotlib inline
df.Close.plot()
```
