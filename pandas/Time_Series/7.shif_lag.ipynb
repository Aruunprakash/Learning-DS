```python
import pandas as pd
```

```python
df = pd.read_csv("fb.csv",parse_dates=["Date"],index_col="Date")
df
```

shifting prices

```python
df.shift(+1)
```

```python
df['prev_day']=df.shift(-1)
df
```

```python
df['price_change']= df['Price'] - df['prev_day']
df
```

```python
df['5 day return'] =  (df['Price'] - df['Price'].shift(5))*100/df['Price'].shift(5)
df
```

changing date

```python
df = df[['Price']]
df
```

```python
df.index = pd.date_range(start="2017-08-15",periods=10,freq="B")
df.index
```

adjust the date

```python
df.shift(periods=1, freq='D')
```
