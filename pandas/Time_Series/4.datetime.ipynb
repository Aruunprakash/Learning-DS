```python
import pandas as pd
```

```python
dates = ['2017-01-05 02:30:00 PM', 'Jan 5, 2017', '01/05/2017', '2017.01.05 01:30:00 AM', '2017/01/05','20170105']
pd.to_datetime(dates,format="mixed")
```

```python
pd.to_datetime('5/1/2017',dayfirst=True)
```

Custom date time format

```python
pd.to_datetime('5$1$2017',format="%d$%m$%Y") #big y for reading 4 digit small digit means it only reads 20 in 2017
```

error handling

```python
dates = ['2017-01-05 02:30:00 PM', 'Jan 5, 2017', '01/05/2017', '2017.01.05 01:30:00 AM', '2017/01/05','20170105','abc']
pd.to_datetime(dates,format="mixed",errors="coerce")
```

Epoch or Unix time means number of seconds that have passed since Jan 1, 1970 00:00:00 UTC time

```python
epoch = 1501324478
dt = pd.to_datetime(epoch, unit='s')
dt
```
