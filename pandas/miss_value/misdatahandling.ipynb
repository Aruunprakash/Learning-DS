```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv",parse_dates=['day'])
type(df.day[0])
df
```

```python
df.set_index('day',inplace=True) # inplace is to apply instead of copying
df
```

fillna with one value for all

```python
new_df = df.fillna(0)
new_df
```

Fill na using column names and dict

```python
new_df = df.fillna({
        'temperature': 0,
        'windspeed': 0,
        'event': 'No Event'
    })
new_df
```

FFILL


```python
new_df= df.ffill()
```

```python
new_df
```

```python
new_df=df.bfill()
new_df
```

Use of axis

```python
new_df = df.bfill(axis='columns')
new_df
```

limit parameter

```python
new_df = df.ffill(limit=1)
```

```python
new_df
```

interpolate

```python
num = ['temperature','windspeed']
df[num]=df[num].interpolate(method='linear') # parameters for accuracy
```

```python
df['event'] = df['event'].ffill(limit=1)
```

```python
df
```

dropna

```python
new_df = df.dropna()
new_df
```

```python
new_df = df.dropna(how='all')
new_df
```

```python
new_df = df.dropna(thresh=1)
new_df
```

```python
dt = pd.date_range("01-01-2017","01-11-2017")
idx = pd.DatetimeIndex(dt)
df.reindex(idx)
```
