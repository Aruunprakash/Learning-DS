```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("weather_data_replace.csv")
df
```

```python
new_df=df.replace(-99999,np.nan)
new_df
```

multiple values use list

```python
new_df = df.replace([-99999,-88888,0,"0"],np.nan)
new_df
```

use of dict

```python
new_df = df.replace({
    'temperature':-99999,
    'windspeed':[-99999,-88888],
    'event':'0'}
    ,np.nan)
```

```python
new_df
```

mapping with dictionary


```python
new_df = df.replace({
    -99999:np.nan,
    "no event":"Sunny"
})
```

```python
new_df
```

regex- regular expression

```python
new_df = df.replace('[A-za-z]','',regex=True)
new_df
```

this removed all characters

```python
new_df = df.replace({
    'temperature':['A-Za-z'],
    'windspeed':['A-Za-z']}
    ,'',regex=True
)
new_df
```

Replacing list with another list


```python
df = pd.DataFrame({
    'score': ['exceptional','average', 'good', 'poor', 'average', 'exceptional'],
    'student': ['rob', 'maya', 'parthiv', 'tom', 'julian', 'erica']
})
df
```

```python
df.replace(['poor', 'average', 'good', 'exceptional'], [1,2,3,4])
```
