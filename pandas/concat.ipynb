```python
import pandas as pd
```

```python
weather_india = pd.DataFrame({
    'city':['mumbai','delhi','bangalore'],
    'temperature':[32,45,30],
    'humidity':[80,60,78]
})
```

```python
weather_india
```

```python
weather_usa = pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35],
    'humidity':[68,65,75]
})
```

```python
weather_usa
```

merge two dataframes

```python
df = pd.concat([weather_india,weather_usa],ignore_index=True) #two avoid og index
```

```python
df
```

passing keys

```python
df = pd.concat([weather_india,weather_usa],keys=['india','usa'])
df
```

```python
df.loc['usa']
```

```python
temp_data=pd.DataFrame({
    'city':['new york','chicago','orlando'],
    'temperature':[21,14,35]
})
temp_data
```

```python
windspeed=pd.DataFrame({
    'city':['chicago','new york'],
    'temperature':[7,12]
},index=[1,0]) #for having proper order
windspeed
```

concat as rows


```python
df=pd.concat([temp_data,windspeed],axis=1) #rows
df
```

```python
temp_data
```

join dataframe with series

```python
s = pd.Series(["humid","dry","rain"],name="event")
```

```python
df = pd.concat([temp_data,s],axis=1)
df
```
