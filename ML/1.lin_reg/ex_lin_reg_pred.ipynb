```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df.head()
```

```python
reg = linear_model.LinearRegression() #model defined
```

```python
df.columns
```

```python
plt.scatter(df.year,df['per capita income (US$)'])
```

```python
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
find_year = pd.DataFrame({
    'year': [2000]
})
```

```python
reg.predict(find_year)
```

```python
plt.xlabel("Year")
plt.ylabel("Percent Income (US$)")
plt.title("Percent Income vs Year")
plt.scatter(df.year,df['per capita income (US$)'],label="Percent Income (US$)")
plt.plot(df.year,reg.predict(df[['year']]),color='red',label='Linear Regression')
plt.legend()
plt.legend()
```
