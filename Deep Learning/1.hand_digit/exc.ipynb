```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
import numpy as np
```

```python
(X_train, y_train), (X_test, y_test) = keras.datasets.mnist.load_data()
```

scaling

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_flatten = X_train.reshape(len(X_train),28*28)
X_flatten
```

```python
X_test_flatten = X_test.reshape(len(X_test),28*28)
X_test_flatten
```

```python
(X_test_flatten[0])
```

Defining neural network

NO HIDDEN LAYER

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,),activation='sigmoid') #10 is output layer , 784 is input layer , then activation fn
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(X_flatten,y_train, epochs=5)
```

Accuracy increased after scaling

```python
model.evaluate(X_test_flatten,y_test)
```

```python
plt.matshow(X_test[0])
```

```python
y_pred = model.predict(X_test_flatten)
y_pred[0]
```

```python
np.argmax(y_pred[0])
```

```python
plt.matshow(X_test[1])
```

```python
y_pred[1]
```

```python
y_pred_labels = [np.argmax(i) for i in y_pred]
y_pred_labels[:5]
```

```python
y_test[:5]
```

```python
cm = tf.math.confusion_matrix(labels=y_test, predictions=y_pred_labels)
cm
```

```python
import seaborn as sn
plt.figure(figsize=(7,5))
sn.heatmap(cm,annot=True,fmt='d')
plt.show()
plt.xlabel('Predicted')
plt.ylabel('True')
```

with hidden layer

```python
model = keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'), # hidden layer , then input ,activation_fun
    keras.layers.Dense(10,activation='sigmoid') #output layer
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(X_flatten,y_train,epochs=5)
```

without hidden layer, the accuracy was 92, no its 97

```python
model.evaluate(X_test_flatten,y_test)
```

```python
y_predicted = model.predict(X_test_flatten)
y_predicted_labels = [np.argmax(i) for i in y_predicted]
cm = tf.math.confusion_matrix(labels=y_test,predictions=y_predicted_labels)

plt.figure(figsize = (8,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

Flatten built-in function

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)), #input layer
    keras.layers.Dense(100,activation='relu'), # hidden layer
    keras.layers.Dense(10,activation='sigmoid') #output layer
])
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```
