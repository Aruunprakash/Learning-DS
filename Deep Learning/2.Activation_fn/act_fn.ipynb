```python
import math
```

```python
def sigmoid(x):
    return 1 / (1 + math.exp(-x))
```

```python
sigmoid(0.5)
```

```python
sigmoid(-0.5)
```

```python
sigmoid(0)
```

```python
def tanh(x):
    return math.tanh(x)
    # (math.exp(x)-math.exp(-x))/ (math.exp(x) + math.exp(-x))
```

```python
tanh(-34)
```

```python
tanh(78)
```

```python
tanh(7)
```

```python
def relu(x):
    return max(0,x)
```

```python
relu(-7)
```

```python
relu(8)
```

```python
relu(0)
```

```python
def leaky_relu(x):
    return max(0.1*x,x)
```

```python
leaky_relu(123)
```

```python
leaky_relu(-90)
```

```python
leaky_relu(0)
```
