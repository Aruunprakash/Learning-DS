FILTERING:-select elements on basis of given condition

```python
ages=np.array([[21,17,19,20,16,30,18,65],
              [39,22,15,99,18,19,20,21]])
```

```python
teen=ages[ages<18]
```

```python
teen #boolean indexing
```

& and , | or

```python
adults=ages[(ages>=18) | (ages<65)]
```

```python
adults
```

```python
even=ages[ages%2==0]
even
```

where(condition,array,fillvalue)


```python
adults=np.where(ages>=18,ages,0)
```

```python
adults
```

```python
adults.shape
```
